```python
# Calculating skewness and kurtosis for the 'speed' and 'distance' columns
speed_skew = skew(Data1['speed'])
distance_skew = skew(Data1['dist'])
```

```python

speed_kurtosis = kurtosis(Data1['speed'])
distance_kurtosis = kurtosis(Data1['dist'])
```

```python
# Interpretation
# Skewness:
# - If skewness is close to 0, the data is approximately symmetric.
# - If skewness is negative, the data is skewed left [tail on the left side].
# - If skewness is positive, the data is skewed right [tail on the right side].

# Kurtosis:
# - If kurtosis is close to 0, the distribution has tails similar to a normal distribution (mesokurtic).
# - If kurtosis is less than 0, the distribution is platykurtic (light tails).
# - If kurtosis is greater than 0, the distribution is leptokurtic (heavy tails).

# Based on the skewness and kurtosis values, you can draw inferences about the shape of the distributions
#of the data.
```

```python
#  for Printing skewness and kurtosis values
print("Skewness of Speed:", speed_skew)
print("Skewness of Distance:", distance_skew)
print("Kurtosis of Speed:", speed_kurtosis)
print("Kurtosis of Distance:", distance_kurtosis)
```

Question no - 22 Calculate the Z scores of  90% confidence interval,94% confidence interval, 60% confidence interval

Question no - 23 Calculate the t scores of 95% confidence interval, 96% confidence interval, 99% confidence interval for sample size of 25

```python
import scipy.stats as stats

# Degrees of freedom
df = 25 - 1

# 95% Confidence Interval (97.5th percentile)
t_95 = stats.t.ppf(0.975, df)

# 96% Confidence Interval (98th percentile)
t_96 = stats.t.ppf(0.98, df)

# 99% Confidence Interval (99.5th percentile)
t_99 = stats.t.ppf(0.995, df)

print("t-score for 95% Confidence Interval:", t_95)
print("t-score for 96% Confidence Interval:", t_96)
print("t-score for 99% Confidence Interval:", t_99)
```

Question no -24 A Government  company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing. The sampled bulbs last an average of 260 days, with a standard deviation of 90 days. If the CEO's claim were true, what is the probability that 18 randomly selected bulbs would have an average life of no more than 260 days

```python
import scipy.stats as stats

# Given data
mu = 270  # Population mean
x_bar = 260  # Sample mean
s = 90  # Sample standard deviation
n = 18  # Sample size

# Calculate the t-score
# Calculate the t-score
t_score = (x_bar - mu) / (s / (n ** 0.5))

# Calculate the probability using the cumulative distribution function (CDF)
probability = stats.t.cdf(t_score, df=n-1)

print("t-score:", t_score)
print("Probability:", probability)
```

```python
import numpy as np

# Data
x_values = [-2000, -1000, 0, 1000, 2000, 3000]
probabilities = [0.1, 0.1, 0.2, 0.2, 0.3, 0.1]

# Calculating expected value (mean)
mean = np.dot(x_values, probabilities)
print("Mean (Expected value):", mean)

# Calculating standard deviation
variance = np.dot((np.array(x_values) - mean)**2, probabilities)
std_dev = np.sqrt(variance)
print("Standard Deviation:", std_dev)
```

Question no - 5

Returns on a certain business venture, to the nearest $1,000, are known to follow the following probability distribution

 (i)	What is the most likely monetary outcome of the business venture?
(ii)	Is the venture likely to be successful? Explain
(iii)	What is the long-term average earning of business ventures of this kind? Explain
(iv)	What is the good measure of the risk involved in a venture of this kind? Compute this measure



```python

returns = [-2000, -1000, 0, 1000, 2000, 3000]
probabilities = [0.1, 0.1, 0.2, 0.2, 0.3, 0.1]
```

```python
expected_return = sum(r * p for r, p in zip(returns, probabilities))
print(f"Expected Return: {expected_return}")
variance = sum((r - expected_return) ** 2 * p for r, p in zip(returns, probabilities))
std_dev = variance ** 0.5
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```
