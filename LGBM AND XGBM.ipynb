```python
df = pd.read_csv("Titanic_train.csv")
df.head()
```

```python
## so first i am doing data transformation categorical to numerical
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
  df = df.drop(columns='Name')
```

```python
print(df.columns)
```

```python
df['PassengerId'] = le.fit_transform(df['PassengerId'])
df['Survived'] = le.fit_transform(df['Survived'])
df['Pclass'] = le.fit_transform(df['Pclass'])
df['SibSp'] = le.fit_transform(df['SibSp'])
df['Parch'] = le.fit_transform(df['Parch'])
df['Fare'] = le.fit_transform(df['Fare'])
df['Ticket'] = le.fit_transform(df['Ticket'])
df['Cabin'] = le.fit_transform(df['Cabin'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df['sex'] = le.fit_transform(df['Sex'])
df
```

```python
df
```

```python
## Univariate Analysis
# Calculate summary statistics
mean = df['Age'].mean()
median = df['Age'].median()
std_dev = df['Age'].std()

# Create frequency table
frequency_table = df['Age'].value_counts()

print('mean:',mean)
print("median:",median)
print("standrad dev:",std_dev)
print(frequency_table)
```

```python

## Bivariate Analysis
import pandas as pd
import matplotlib.pyplot as plt
#scatter plot
plt.scatter(df.Age, df.Fare)
plt.title('Sex vs. Fare ')
plt.xlabel('Sex')
plt.ylabel('Fare')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df)
plt.show()
```
